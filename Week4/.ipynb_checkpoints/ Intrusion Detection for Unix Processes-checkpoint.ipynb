#  Intrusion Detection for Unix Processes

## Part 1: Preprocessing cert

First, we need to convert the training data to a file with only fixed sized substring.

```python
import numpy as np
import sklearn.metrics as metrics
import matplotlib.pyplot as plt
import collections
```

```python
train = np.loadtxt("syscalls/snd-cert/snd-cert.train", dtype=str)

def fixed_length_substring(file, K):
    train_substrings = []
    for train_str in file:
        for train_substr in range(0, len(train_str), K):
            new_substr = train_str[0+train_substr:train_substr+K]
            if len(new_substr) == K:
                train_substrings.append(new_substr)
    return list(set(train_substrings))
                
new_train = fixed_length_substring(train, 7) #this will perfectly fit AEEEEE

def write_file(l, name):
    new_str = ""
    for i in l:
        new_str = new_str + i + "\n"

    file = open(name, "w+")
    file.write(new_str)
    file.close()

write_file(new_train, "snd-cert-chunk.train")
```

```python
test1 = np.loadtxt("syscalls/snd-cert/snd-cert.1.test", dtype=str)
labels1 = np.loadtxt("syscalls/snd-cert/snd-cert.1.labels", dtype=str)

def fixed_length_substring_with_labels_a(labels, K):
    test_substrings = []
    for i in range(0, len(labels)): 
        for test_substr in range(0, len(labels[i]), K):
            new_substr = labels[i][0+test_substr:test_substr+K]
            if len(new_substr) == K:
                test_substrings.append([new_substr, i])
    a = list(zip(*test_substrings))[0]
    b = list(zip(*test_substrings))[1]
    d = collections.OrderedDict.fromkeys(zip(a,b))
    return [k[0] for k in d], [k[1] for k in d]

new_test1a, new_labels1a = fixed_length_substring_with_labels_a(test1, K)
write_file(new_test1a, "snd-cert-chunka.1.test")
```

```python
test2 = np.loadtxt("syscalls/snd-cert/snd-cert.2.test", dtype=str)
labels2 = np.loadtxt("syscalls/snd-cert/snd-cert.2.labels", dtype=str)
new_test2a, new_labels2a = fixed_length_substring_with_labels_a(test2, K)
write_file(new_test2a, "snd-cert-chunka.2.test")

test3 = np.loadtxt("syscalls/snd-cert/snd-cert.3.test", dtype=str)
labels3 = np.loadtxt("syscalls/snd-cert/snd-cert.3.labels", dtype=str)
new_test3a, new_labels3a = fixed_length_substring_with_labels_a(test3, K)
write_file(new_test3a, "snd-cert-chunka.3.test")
```

## Part 2: Classification cert

Now that we have the correct train and test files, we can run the following commands:

`java -jar negsel2.jar -alphabet file://snd-cert.alpha -self snd-cert-chunk.train -n 6 -r 4 -c -l < snd-cert-chunka.1.test > test1.txt`

`java -jar negsel2.jar -alphabet file://snd-cert.alpha -self snd-cert-chunk.train -n 6 -r 4 -c -l < snd-cert-chunka.2.test > test2.txt`

`java -jar negsel2.jar -alphabet file://snd-cert.alpha -self snd-cert-chunk.train -n 6 -r 4 -c -l < snd-cert-chunka.3.test > test3.txt`

compute the number of matching patterns for each chunk separately, and merge these counts together to a composite anomaly score (for instance, you could average the individual counts)

```python
#HEEL GEK IK KRIJG 2 GETALLEN IN MIJN TESTFILES
test1 = [x for [x,y] in np.loadtxt("test1.txt", dtype=float)]
test2 = [x for [x,y] in np.loadtxt("test2.txt", dtype=float)]
test3 = [x for [x,y] in np.loadtxt("test3.txt", dtype=float)]
```

```python
#op de een of andere manier terug vinden wat waarbij hoort
def compute_anomaly_score(labels, new_labels, test):
    anomaly_score_per_string = []
    for i in range(0, len(labels)):
        count_per_string = 0
        substr_in_str = 0
        for j in range(0, len(new_labels)):
            if i == new_labels[j]:
                count_per_string += test[j]
                substr_in_str += 1
        anomaly_score_per_string.append(count_per_string / substr_in_str)
    return anomaly_score_per_string
```

```python
anomaly_scores1 = compute_anomaly_score(labels1, new_labels1a, test1)
anomaly_scores2 = compute_anomaly_score(labels2, new_labels2a, test2)
anomaly_scores3 = compute_anomaly_score(labels3, new_labels3a, test3)
```

```python
def compute_roc_auc(test, labels):
    labels = [int(x) for x in labels] 
    anomaly_scores = list(zip(test, labels))
    anomaly_scores.sort(reverse=True)
    
    #split the list up into actual class (y) and anomal score (y_pred)
    y = [y for [x,y] in anomaly_scores]
    y_pred = [x for [x,y] in anomaly_scores]

    #using the functions from sklearn to easily calculate the roc_curve
    fpr, tpr, threshold = metrics.roc_curve(y, y_pred)
    roc_auc = metrics.auc(fpr, tpr)
    return fpr, tpr, threshold, roc_auc

def plot_ROC(fpr, tpr, roc_auc):
    #plotting the RCO
    plt.title('Receiver Operating Characteristic')
    plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
    plt.legend(loc = 'lower right')
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()
```

```python
fpr, tpr, threshold, roc_auc = compute_roc_auc(anomaly_scores1, labels1)
plot_ROC(fpr, tpr, roc_auc)
```

```python
fpr, tpr, threshold, roc_auc = compute_roc_auc(anomaly_scores2, labels2)
plot_ROC(fpr, tpr, roc_auc)
```

```python
fpr, tpr, threshold, roc_auc = compute_roc_auc(anomaly_scores3, labels3)
plot_ROC(fpr, tpr, roc_auc)
```
